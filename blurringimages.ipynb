```python
import cv2
from matplotlib import pyplot

face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
image = cv2.imread("group.jpg")  # Converts RGB to BGR
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  #Converts BGR to Black and white

faces = face_cascade.detectMultiScale(gray, 1.1, 4)
print(len(faces))  # Prints number of faces

# Make rectangles
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,255), 2)

pyplot.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

```python
import cv2
from matplotlib import pyplot as plt

#load face detector
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

#load face
image = cv2.imread("group.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#detect face
faces = face_cascade.detectMultiScale(gray, 1.1, 4)

#load emoji
emoji = cv2.imread("emoji.png", cv2.IMREAD_UNCHANGED)

if emoji is None:
    raise FileNotFoundError("emoji.png could not be loaded")   #Use as safety checks

if emoji.shape[2] != 4:
    raise ValueError("emoji.png must be RGBA (transparent PNG)")

for (x, y, w, h) in faces:
    emoji_resized = cv2.resize(emoji, (w, h))

    emoji_rgb = emoji_resized[:, :, :3]
    alpha = emoji_resized[:, :, 3] / 255.0

    roi = image[y:y+h, x:x+w]

    for c in range(3):
        roi[:, :, c] = alpha * emoji_rgb[:, :, c] + (1 - alpha) * roi[:, :, c]

    image[y:y+h, x:x+w] = roi

# Show result
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
